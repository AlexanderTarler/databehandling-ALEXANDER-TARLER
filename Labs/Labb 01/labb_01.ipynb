#### Task (f)

In the code below I used both pandas and matplotlib.pyplot to show that the "Elite" colleges are, in general, more expensive when it comes to out-of-state tuition. 

```python
# f

college["Elite"] = pd.cut(college["Top10perc"], [0, 50, 100], labels=["No", "Yes"])

print(college["Elite"].value_counts())

college.boxplot(column='Outstate', by='Elite', grid=False)
plt.title('Outstate Tuition by Elite Status')
plt.suptitle('')  
plt.xlabel('Elite')
plt.ylabel('Outstate Tuition')
plt.show()

```

#### Task (g)

For this task I chose to use the variables "Apps", "Accept", "Enroll" and "Grad.Rate" to see if the data would show any correlation between the four variables. 

```python
# g

variables = ["Apps", "Accept", "Enroll", "Grad.Rate"]
titles = ["Applications (10 bins)", "Acceptances (15 bins)", "Enrollment (20 bins)", "Graduation Rate (25 bins)"]
bins = [10, 15, 20, 25]
x_labels = ["Number of Applications", "Number of Acceptances", "Number of Enrollments", "Graduation Rate (%)"]

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

for i, ax in enumerate(axes.flatten()):
    college[variables[i]].plot.hist(bins=bins[i], ax=ax, title=titles[i])
    ax.set_xlabel(x_labels[i])
    ax.set_ylabel("Number of schools")

plt.tight_layout()
plt.show()
```

```python
# h

variables = ['Outstate', 'Room.Board', 'Books', 'Personal']
titles = ["Out-of-State Tuition (10 bins)", "Room & Board (15 bins)", "Books (20 bins)", "Personal Spending (25 bins)"]
bins = [10, 15, 20, 25]
x_labels = ["Out-of-State Tuition ($)", "Room & Board Costs ($)", "Book Costs ($)", "Personal Spending ($)"]

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

for i, ax in enumerate(axes.flatten()):
    college[variables[i]].plot.hist(bins=bins[i], ax=ax, title=titles[i])
    ax.set_xlabel(x_labels[i])
    ax.set_ylabel("Number of Schools")

plt.tight_layout()
plt.show()
```

These insights highlight variations in costs, popularity, and outcomes across institutions. The patterns suggest that certain factors, such as school type (private vs. public) and elite status, influence both costs and enrollment figures, which could impact accessibility and affordability for students.

#### Room & Board

Room and board costs are shown to generally land between $3,500 and $5,500. This shows a somewhat standardized cost for room and board across all schools, which could reflect on both stable housing and food prizes as being affordable.

#### Book Costs

Book costs are strongly right-skewed (meaning that most data lands on the left side of the graph), with most institutions reporting expenses below $500 a year, but a few outliers spend significantly more. 

This variation might be due to differences in programs and required materials - for instance, institutions offering specialized or resource-heavy programs could have higher book costs.

#### Personal Spending

Personal spending is also right-skewed, with a high frequency of schools reporting around $1,000 in personal expenses, but some outliers reporting up to $7,000. The outliers might be due to differences in costs between urban and rural schools, where prize of housing and other expenses can differ greatly.


#### Relationship Between Costs and Elite Status

Elite institutions (those with a higher percentage of students from the top 10% of their high school class) tend to have higher out-of-state tuition costs, as observed in previous boxplots. This suggests that elite institutions may use higher tuition as a premium, reflecting their reputation or exclusivity. This might be a factor in increasing the gap between the richer classes with the less fortunate ones, since cost could dictate whether or not you get educated.

Elite institutions may also have higher personal spending expectations, either due to campus amenities or due to their locations in urban areas, which often have a higher cost of living.
